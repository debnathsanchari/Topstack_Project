Project: Health Data Analysis

Problem Definition: Analyze synthetic health data to identify trends.

To-Do List:

Generate synthetic health data.
Calculate summary statistics for health measurements.
Create line plots to visualize health trends over time.
Column Names:

date: Date of health data
measurement: Health measurement (e.g., weight, BMI)
value: Measurement value

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
from datetime import datetime, timedelta

# Generate synthetic health data
np.random.seed(0)  # Set a seed for reproducibility
num_records = 365  # Number of days of data

# Create random date range for one year
start_date = datetime(2022, 1, 1)
end_date = start_date + timedelta(days=num_records - 1)
date_range = [start_date + timedelta(days=i) for i in range(num_records)]

# Create synthetic health data with random measurements
data = {
    'date': [],
    'measurement': [],
    'value': [],
}

for date in date_range:
    for measurement in ['weight', 'BMI']:
        data['date'].append(date)
        data['measurement'].append(measurement)
        data['value'].append(random.uniform(50, 100))

# Convert data to a DataFrame
df = pd.DataFrame(data)

# Calculate summary statistics
summary_stats = df.groupby('measurement')['value'].describe()

# Create line plots to visualize health trends over time
plt.figure(figsize=(12, 8))

for measurement in df['measurement'].unique():
    measurement_data = df[df['measurement'] == measurement]
    plt.plot(measurement_data['date'], measurement_data['value'], label=measurement , marker='o', linestyle='-')

# Display summary statistics
print(summary_stats)

plt.title('Health Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Measurement Value')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
